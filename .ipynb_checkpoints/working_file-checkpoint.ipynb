```python
with urllib.request.urlopen("https://api.census.gov/data/2018/acs/acs1?get=NAME,B01001_001E&for=state:*") as url:
    state_data = json.loads(url.read().decode())
```

```python
state_pop_dict = {us_state_abbrev[i[0]]: int(i[1]) for i in state_data[1:]}
    
```

```python
df = pd.read_csv("https://covidtracking.com/api/v1/states/daily.csv")[['date', 'state', 'positive', 'negative', 'death']]
df.date = pd.to_datetime(df.date, format='%Y%m%d')

df = df.fillna(0)
df.head()
```

```python
state_dict = []

for state in df.state.unique():

    dict_ = df[df.state == state].sort_values('date', ascending= False).to_dict('records')
    period = 7
# Calculating the 
    for i in range(0, len(dict_)-period):
        pos = sum([dict_[j]['positive'] for j in range(i, i+ period)])
        dict_[i]['change'] = (dict_[i]['positive'] - dict_[i+1]['positive'])
        dict_[i]['weekly_change'] = (dict_[i]['positive'] - dict_[i+7]['positive'])/ dict_[i+7]['positive']
        neg = sum([dict_[j]['negative'] for j in range(i, i+ period)])
        if (neg+ pos) > 0:
            dict_[i]['rate'] = pos/ (neg+ pos)
        else:
            dict_[i]['rate'] = 0
        if state not in ['AS', 'GU', 'MP', 'VI']:
            dict_[i]['test_per_1000'] = (neg+ pos)* 1000/ (state_pop_dict[state]* 7)
            state_dict.append(dict_[i])
        
```

```python
df = pd.DataFrame(state_dict)
df.head()
```

```python
ax=df.loc[(df.date > (datetime.datetime.today() - datetime.timedelta(days=30))) & (df.state == 'FL')].plot(x = 'date', y = 'death');
ax1 = df.loc[(df.date > (datetime.datetime.today() - datetime.timedelta(days=30))) & (df.state == 'FL')].plot(x = 'date', y = 'rate');
df.loc[(df.date > (datetime.datetime.today() - datetime.timedelta(days=30))) & (df.state == 'FL')].plot(x = 'date', y = 'change', ax=ax);
df.loc[(df.date > (datetime.datetime.today() - datetime.timedelta(days=30))) & (df.state == 'FL')].plot(x = 'date', y = 'weekly_change', ax=ax1);
```
