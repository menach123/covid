```python
df = pd.read_csv("https://covidtracking.com/api/v1/states/daily.csv")[['date', 'state', 'positive', 'negative', 'death']]
df.date = pd.to_datetime(df.date, format='%Y%m%d')

df = df.fillna(0)


```

```python
state_dict = []

for state in df.state.unique():

    dict_ = df[df.state == state].sort_values('date', ascending= False).to_dict('records')
    period = 7
    for i in range(0, len(dict_)-period):
        pos = sum([dict_[j]['positive'] for j in range(i, i+ period)])
        dict_[i]['change'] = (dict_[i]['positive'] - dict_[i+1]['positive'])
        neg = sum([dict_[j]['negative'] for j in range(i, i+ period)])
        if (neg+ pos) > 0:
            dict_[i]['rate'] = pos/ (neg+ pos)
        else:
            dict_[i]['rate'] = 0
        if state not in ['AS', 'GU', 'MP', 'VI']:
            dict_[i]['test_per_1000'] = (neg+ pos)* 1000/ state_pop_dict[state]   
            state_dict.append(dict_[i])
```
